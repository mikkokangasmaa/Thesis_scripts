```python
# Summing all counter stations together and returning the same DataFrame structure
laskuri_total = pd.DataFrame(laskuri[['2021-01', '2021-02', '2021-03', '2021-04', '2021-05',
       '2021-06', '2021-07', '2021-08', '2021-09', '2021-10', '2021-11',
       '2021-12', '2022-01', '2022-02', '2022-03', '2022-04', '2022-05',
       '2022-06', '2022-07', '2022-08', '2022-09', '2022-10', '2022-11',
       '2022-12', '2023-01', '2023-02', '2023-03', '2023-04', '2023-05',
       '2023-06', '2023-07', '2023-08', '2023-09', '2023-10', '2023-11',
       '2023-12']].sum(axis=0),columns=['total'])

strava_total = pd.DataFrame(strava[['2021-01', '2021-02', '2021-03', '2021-04', '2021-05',
       '2021-06', '2021-07', '2021-08', '2021-09', '2021-10', '2021-11',
       '2021-12', '2022-01', '2022-02', '2022-03', '2022-04', '2022-05',
       '2022-06', '2022-07', '2022-08', '2022-09', '2022-10', '2022-11',
       '2022-12', '2023-01', '2023-02', '2023-03', '2023-04', '2023-05',
       '2023-06', '2023-07', '2023-08', '2023-09', '2023-10', '2023-11',
       '2023-12']].sum(axis=0),columns=['total'])

laskuri_kesa = pd.DataFrame(laskuri[[
       '2021-06', '2021-07', '2021-08', '2021-09', '2021-10', 
       '2022-06', '2022-07', '2022-08', '2022-09', '2022-10', 
       '2023-06', '2023-07', '2023-08', '2023-09', '2023-10']].sum(axis=0),columns=['total'])

strava_kesa = pd.DataFrame(strava[[
       '2021-06', '2021-07', '2021-08', '2021-09', '2021-10', 
       '2022-06', '2022-07', '2022-08', '2022-09', '2022-10', 
       '2023-06', '2023-07', '2023-08', '2023-09', '2023-10']].sum(axis=0),columns=['total'])

laskuri_yllas_kesa = pd.DataFrame(laskuri_yllas[[
       '2021-06', '2021-07', '2021-08', '2021-09', '2021-10', 
       '2022-06', '2022-07', '2022-08', '2022-09', '2022-10', 
       '2023-06', '2023-07', '2023-08', '2023-09', '2023-10']].sum(axis=0),columns=['total'])

strava_yllas_kesa = pd.DataFrame(strava_yllas[[
       '2021-06', '2021-07', '2021-08', '2021-09', '2021-10', 
       '2022-06', '2022-07', '2022-08', '2022-09', '2022-10', 
       '2023-06', '2023-07', '2023-08', '2023-09', '2023-10']].sum(axis=0),columns=['total'])
strava_total.head()

strava_kesa == strava_yllas_kesa
```

```python
print(strava_yllas_kesa)
print(strava_kesa)
```

```python
def prop (row):
    proportion = (row/row.sum())*100
    return proportion

laskuri_total['prop'] = laskuri_total.apply(prop,axis=0)
strava_total['prop'] = strava_total.apply(prop,axis=0)

laskuri_kesa['prop'] = laskuri_kesa.apply(prop,axis=0)
strava_kesa['prop'] = strava_kesa.apply(prop,axis=0)

laskuri_yllas_kesa['prop'] = laskuri_yllas_kesa.apply(prop,axis=0)
strava_yllas_kesa['prop'] = strava_yllas_kesa.apply(prop,axis=0)

#strava_yllas_kesa == strava_kesa
```

```python
strava_kesa == strava_yllas_kesa
```

```python
kaikki_total = pd.merge(laskuri_total,strava_total,left_index=True,right_index = True, suffixes=('_lask','_strava'))

kaikki_kesa = pd.merge(laskuri_kesa,strava_kesa,left_index=True,right_index = True, suffixes=('_lask','_strava'))

kaikki_kesa_yllas = pd.merge(laskuri_yllas_kesa,strava_yllas_kesa,left_index=True,right_index = True, suffixes=('_lask','_strava'))


kaikki_kesa_yllas == kaikki_kesa
```
