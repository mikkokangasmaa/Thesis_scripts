```python
def prop (row):
    proportion = (row/row.sum())*100
    return proportion

laskuri_yllas_total['prop_yllas_l'] = laskuri_yllas_total.apply(prop,axis=0)
strava_yllas_total['prop_yllas_s'] = strava_yllas_total.apply(prop,axis=0)

laskuri_yllas_kesa['prop_yllas_l'] = laskuri_yllas_kesa.apply(prop,axis=0)
strava_yllas_kesa['prop_yllas_s'] = strava_yllas_kesa.apply(prop,axis=0)


laskuri_pyha_total['prop_pyha_l'] = laskuri_pyha_total.apply(prop,axis=0)
strava_pyha_total['prop_pyha_s'] = strava_pyha_total.apply(prop,axis=0)

laskuri_pyha_kesa['prop_pyha_l'] = laskuri_pyha_kesa.apply(prop,axis=0)
strava_pyha_kesa['prop_pyha_s'] = strava_pyha_kesa.apply(prop,axis=0)


laskuri_uk_total['prop_uk_l'] = laskuri_uk_total.apply(prop,axis=0)
strava_uk_total['prop_uk_s'] = strava_uk_total.apply(prop,axis=0)

laskuri_uk_kesa['prop_uk_l'] = laskuri_uk_kesa.apply(prop,axis=0)
strava_uk_kesa['prop_uk_s'] = strava_uk_kesa.apply(prop,axis=0)





#laskuri_yllas_kesa['prop'] = laskuri_yllas_kesa.apply(prop,axis=0)
#strava_yllas_kesa['prop'] = strava_yllas_kesa.apply(prop,axis=0)

#strava_yllas_kesa == strava_kesa
```
