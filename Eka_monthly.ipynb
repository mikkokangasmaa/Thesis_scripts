```python
geo_data.rename(columns={'edgeUID': 'edge_uid'}, inplace=True)

geo_data.head()
```

```python
data = data[['edge_uid','month','total_trip_count']]

data.head()
```

```python
# Group the data by 'edge_uid' and 'month', and sum the 'total_trip_count'
grouped_data = data.groupby(['edge_uid', 'month'])['total_trip_count'].sum().reset_index()

# Pivot the data
pivot_data = grouped_data.pivot_table(index='edge_uid', columns='month', values='total_trip_count', fill_value=0)

# If you want to reset the index and have 'edge_uid' as a column instead
pivot_data.reset_index(inplace=True)




```

```python
# Convert pivot_data to a GeoDataFrame with 'edge_uid' as the index
#pivot_data_gdf = gpd.GeoDataFrame(pivot_data)

# Join pivot_data_gdf to data_geom using the common column 'edge_uid'
joined_gdf = geo_data.merge(pivot_data, on='edge_uid')
joined_gdf.head
```

```python
joined_gdf.to_file('2023_monthly_totaltrip.gpkg')
```
